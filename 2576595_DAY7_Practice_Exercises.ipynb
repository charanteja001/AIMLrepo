# 1.Write a Python program to find a target values in a list using linear search with following steps: 
a. Initialize the list to store the input elements.

b. Initialize found-False.

C. Enter the item to be searched (match_item).

d. For each element in the list

1. if match_item = value

a. return match_item's position.

e. If the match item is not in the list, display an error message that the item is not found in the list.

```python
input_list = [34,89,56,34,76,89,12,34,45,32,100]

found = False

match_item = int(input("Enter the item to search : "))
position = []
for i, value in enumerate(input_list):
    if match_item == value:
        found = True
        position.append(i)

if found:
    print(f"The item {match_item} was found at position {position}.")
else:
    print(f"The item {match_item} is not available.")
```

# 2. Write a Python program to implement binary search to find the target values from the list:
a. Create a separate function to do binary search. 
b. Get the number of inputs from the user. 
c. Store the inputs individually in a list. 
d. In binary search function at first sort the list in order to start the search from middle of the list 
e. Compare the middle element to right and left elements to search target element. 
f. If greater, move to right of list or else move to another side of the list. 
g. Print the result along with the position of the element

```python
def binary(arr, target):
    arr.sort()
    left ,right = 0, len(arr) - 1

    while left <= right:
        mid = (left + right) // 2  

        if arr[mid] == target:
            return f"Target {target} found at position {mid}"
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return f"Target {target} not found in the list"


Inputs = int(input("Enter the number of inputs: "))

input_list = []
for i in range(Inputs):
    input_value = int(input(f"Enter input {i + 1}: "))
    input_list.append(input_value)

target_value = int(input("Enter the target value to search for: "))

result = binary(input_list, target_value)

print(result)
```

# 6. Write a Python script to perform the following operations on a singly linked list¶
a. Create a list 
b. Find the smallest element from the list 
c. Insert an element if it is not a duplicate element 
d. Display the elements in reverse order

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class SinglyLinkedList:
    def __init__(self):
        self.head = None

    def insert(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node

    def find_smallest(self):
        if not self.head:
            return None
        smallest = self.head.data
        current = self.head.next
        while current:
            if current.data < smallest:
                smallest = current.data
            current = current.next
        return smallest
    def insert_if_not_duplicate(self, data):
        current = self.head
        while current:
            if current.data == data:
                return  
            current = current.next
        self.insert(data)

    def display_reverse(self):
        def display_recursive(node):
            if node:
                display_recursive(node.next)
                print(node.data, end=' ')

        display_recursive(self.head)

my_list=SinglyLinkedList()
my_list.insert(4)
my_list.insert(54)
my_list.insert(22)
my_list.insert(40)

smallest = my_list.find_smallest()
if smallest is not None:
    print(f"Smallest element in the list: {smallest}")
else:
    print("The list is empty")

my_list.insert_if_not_duplicate(67)
print("Elements in reverse order:")
my_list.display_reverse()
```

# 7.Write a python program to implement the various operations for stack ADT

i.) Push      ii.) Pop     iii.) Display


```python
class Stack():
    
    def __init__(self):
        self.__stack = []
        
    def isEmpty(self):
        if len(self.__stack)==0:
            return True 
        else:
            return False
        
    def Push(self, data):
        self.__stack.append(data)
        print(f"Pushed {data} into the stack.")
        
    def Pop(self):
        if len(self.__stack)==0:
            raise Exception("Stack is empty")
        
        else:
            popped_data = self.__stack.pop()
            print(f"Popped {popped_data} from the stack.")
            return popped_data
        
    def Display(self):
        if len(self.__stack)==0:
            raise Exception("stack is empty")
        else:
            print("Displaying Stack : ")
            for item in self.__stack[::-1]:
                print(item)
        
        
        
        
        
s = Stack()
 
s.Push(1)
s.Push(2)
s.Push(3)
s.Push(4)
s.Push(5)

s.Display()


s.Pop()
s.Pop()
s.Pop()

s.Display()
```

# 8. Write a python script to implement the various operations for Queue ADT

i.) Insert              ii.) Delete          iii.) Display

```python
class Queue():
    
    def __init__(self):
        self.__queue = []
        
    def isEmpty(self):
        if len(self.__queue)==0:
            return True 
        else:
            return False
        
    def Enqueue(self, data):
        self.__queue.append(data)
        print(f"Inserted {data} into the Queue.")
        
    def Dequeue(self):
        if len(self.__queue)==0:
            raise Exception("Queue is empty")
        
        else:
            deleted_data = self.__queue.pop(0)
            print(f"Deleted {deleted_data} from the Queue.")
            return deleted_data
        
    def Display(self):
        if len(self.__queue)==0:
            raise Exception("Queue is empty")
        else:
            print("Displaying Queue : ")
            for item in self.__queue[::-1]:
                print(item)
        
        
        
        
        
q = Queue()
 
q.Enqueue(5)
q.Enqueue(10)
q.Enqueue(15)
q.Enqueue(4)
q.Enqueue(96)

q.Display()


q.Dequeue()
q.Dequeue()


q.Display()


```

# 9. Write a program in python to convert the following infix expression to its postfix form using push and pop operations of a Stack
a. A/B^C+DE-FG

b. (B^2-4AC)^(1/2) (100)

```python
class Stack:
    def __init__(self):
        self.items = []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()
        else:
            return None

    def is_empty(self):
        return len(self.items) == 0

    def peek(self):
        if not self.is_empty():
            return self.items[-1]
        else:
            return None

def infix_to_postfix(expression):
    precedence = {'+': 1, '-': 1, '*': 2, '/': 2, '^': 3}
    output = []
    stack = Stack()

    for char in expression:
        if char.isalnum():
            output.append(char)
        elif char == '(':
            stack.push(char)
        elif char == ')':
            while not stack.is_empty() and stack.peek() != '(':
                output.append(stack.pop())
            stack.pop()  
        else:
            while not stack.is_empty() and precedence.get(char, 0) <= precedence.get(stack.peek(), 0):
                output.append(stack.pop())
            stack.push(char)
    while not stack.is_empty():
        output.append(stack.pop())

    return ''.join(output)

infix_expr_a = "A/B°C+D*E-F*G"
postfix_a = infix_to_postfix(infix_expr_a)
print("Postfix expression for a:", postfix_a)

infix_expr_b = "(B^2-4*A*C)^(1/2) (100)"
postfix_b = infix_to_postfix(infix_expr_b)
print("Postfix expression for b:", postfix_b)
```
